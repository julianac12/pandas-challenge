# School Summary

```python
school_data_complete
```

```python
school_type = school_data_complete.groupby(["school_name"])["type"].first()
school_type
```

```python
student_per_school = school_data_complete["school_name"].value_counts()
student_per_school
```

```python
budget_per_school = school_data_complete.groupby(["school_name"])["budget"].mean()
```

```python
budget_per_school
```

```python
budget_per_student = budget_per_school/student_per_school
```

```python
budget_per_student
```

```python
average_math_score = school_data_complete.groupby(["school_name"])["math_score"].mean()
```

```python
average_math_score
```

```python
average_reading_score = school_data_complete.groupby(["school_name"])["reading_score"].mean()
```

```python
average_reading_score
```

```python
passing_math_score = school_data_complete["math_score"]>=70
```

```python
percent_pass_math_per_school = passing_math_score.groupby(school_data_complete["school_name"]).mean()*100
```

```python
percent_pass_math_per_school
```

```python
passing_reading_score = school_data_complete["reading_score"]>=70
```

```python
percent_pass_reading_per_school = passing_reading_score.groupby(school_data_complete["school_name"]).mean()*100
```

```python
percent_pass_reading_per_school
```

```python
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
```

```python
school_students_passing_math_and_reading
```

```python
overall_students_passing_math_and_reading = school_students_passing_math_and_reading/ student_per_school *100
```

```python
overall_students_passing_math_and_reading
```

```python
per_school_summary = pd.DataFrame({
    "School Type" : school_type,
    "Total Students" : student_per_school,
    "Total School Budget" : budget_per_school,
    "Per Student Budget": budget_per_student, 
    "Avg Math Score" : average_math_score,
    "Avg Reading Score" : average_reading_score, 
    "% Passing Math" : percent_pass_math_per_school,
    "% passing Reading" : percent_pass_reading_per_school,
    "% Overall Passing" : overall_students_passing_math_and_reading})
per_school_summary
```

```python
per_school_summary_descending = per_school_summary.sort_values(by="% Overall Passing" , ascending = False)
top_schools = per_school_summary_descending.head()
top_schools
```

```python
per_school_summary_ascending = per_school_summary.sort_values(by="% Overall Passing" , ascending = True)
bottom_schools = per_school_summary_ascending.head()
bottom_schools
```

# Scores by School Spending 

```python
pd.DataFrame(budget_per_student)
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(budget_per_student,spending_bins,labels=labels)
school_spending_df
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
spending_summary_df = pd.DataFrame({
        "Avg Math Score" : spending_math_scores,
        "Avg Reading Score" : spending_reading_scores,
        "% Passing Math": spending_passing_math,
        "% Passing Reading": spending_passing_reading, 
        "% Overall Passing": overall_passing_spending})
spending_summary_df
```

# Scores by School Size

```python
pd.DataFrame(student_per_school)
```

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_size_df = per_school_summary.copy()
school_size_df
```

```python
school_size_df["School Size"] = pd.cut(student_per_school,size_bins,labels=labels)
school_size_df
```

```python
size_math_scores = school_size_df.groupby(["School Size"])["Avg Math Score"].mean()
size_reading_scores = school_size_df.groupby(["School Size"])["Avg Reading Score"].mean()
size_passing_math = school_size_df.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = school_size_df.groupby(["School Size"])["% passing Reading"].mean()
size_overall_passing = school_size_df.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
size_summary = pd.DataFrame({
        "Avg Math Score" : size_math_scores,
        "Avg Reading Score" : size_reading_scores,
        "% Passing Math": size_passing_math,
        "% Passing Reading": size_passing_reading, 
        "% Overall Passing": size_overall_passing})
size_summary
```

```python
average_math_score_by_type = per_school_summary.groupby(["School Type"])["Avg Math Score"].mean()
average_reading_score_by_type = per_school_summary.groupby(["School Type"])["Avg Reading Score"].mean()
average_percent_passing_math_by_type = per_school_summary.groupby(["School Type"])["% Passing Math"].mean()
average_percent_passing_reading_by_type = per_school_summary.groupby(["School Type"])["% passing Reading"].mean()
average_percent_overall_passing_by_type = per_school_summary.groupby(["School Type"])["% Overall Passing"].mean()
```

```python
type_summary = pd.DataFrame ({
        "Avg Math Score" : average_math_score_by_type,
        "Avg Reading Score" : average_reading_score_by_type,
        "% Passing Math": average_percent_passing_math_by_type,
        "% Passing Reading": average_percent_passing_reading_by_type, 
        "% Overall Passing": average_percent_overall_passing_by_type})
type_summary
```
